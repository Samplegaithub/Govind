```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(customer_data[features])
```

```python
from sklearn.decomposition import PCA

# Reduce dimensions to 2D
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(customer_data[features])

# Plot clusters
plt.figure(figsize=(8, 6))
for cluster in customer_data['Cluster'].unique():
    cluster_points = reduced_data[customer_data['Cluster'] == cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster}')
plt.title('Customer Clusters')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.show()
```

```python
customer_data_numeric = customer_data.select_dtypes(exclude=['object'])
cluster_summary = customer_data_numeric.groupby('Cluster').mean()
print(cluster_summary)
```
